# Data Cleaning
Visualizing the text-input reveals some messy data issues. We will have to spend some elbow grease to clean that up.

One of the most powerful tools for dealing with text data is **regular expressions**, which we will apply to help clean that data.

```python
df['Gender'] = df['Gender'].replace('f', 'female')
df['Gender'] = df['Gender'].replace('m', 'male')
```

```python
df['Gender'].apply(lambda x: x.lower()).value_counts().plot(kind="bar", color='red')
```

```python
df['Race'].apply(lambda x: x.strip().lower()).value_counts().plot(kind="pie", colormap="plasma")
```

```python
df['Salary']
```

```python
df['Salary'] = df['Salary'].str.replace('$', '')
df['Salary']= df['Salary'].str.replace(',','').astype(float)
df['Salary']
```

# Ethical Computing?

Computer ethics is all about making good choices when using technology. Just like we learn to make good choices in our everyday life, it's important to learn how to use technology in a responsible way.

<img src="https://i0.wp.com/classnotes.ng/wp-content/uploads/2020/04/Computer-Ethics.jpg?fit=1800%2C1200&ssl=1" width="50%"/>

*   **Be respectful:** Treat others with kindness and respect online, just like you would in person.

*   **Protect your privacy:** Keep your personal information safe by not sharing it online. This includes your full name, address, phone number, and other sensitive information. Be careful when sharing pictures, too. Once something is online, it's hard to take it back.

*  **Think before you post:** Before you post something online, ask yourself if it's something you would be proud to share with your parents, teachers, or future employers. Remember that once something is online, it can be seen by anyone.

*  **Follow the rules:** Just like there are rules in the classroom, there are rules online, too. Make sure you know what's acceptable and what's not, and follow the rules set by your school, parents, and other authorities.

*   **Be a good digital citizen:** Remember that everything you do online reflects on you as a person. Be a good digital citizen by using technology responsibly and making positive contributions to the online community.



# How to be an ethical Data Scientist?
Data science is the study of using data to solve problems and make informed decisions. While data can be incredibly useful, it also presents challenges when it comes to using it responsibly. In this lesson, we'll explore ethical data science and discuss some of the ways we can use data in a positive and responsible way.


**What is ethical data science?**

Ethical data science involves using data in a responsible, transparent, and fair way.
This means considering the impact of data on people and society, and taking steps to ensure that data is collected, analyzed, and used in an ethical way.

**Why is ethical data science important?**

Ethical data science is important because data can have a powerful impact on people's lives.
Data can be used to make decisions about everything from healthcare to education to job opportunities.
If data is used in an unfair or biased way, it can lead to harm or discrimination.

**How can we use data responsibly?**

We can use data responsibly by being transparent about how it is collected and used.
We can also take steps to ensure that data is unbiased and representative of the population it represents.
We should also consider the potential impact of data on people's lives and take steps to minimize harm or discrimination.

<img src="https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcTmhg2KkS60GmVMk4UycaAV8jtbcxse-h9hHw&usqp=CAU" width="50%"/>

#### **What is Privacy?**
Privacy is all about keeping your personal information safe and secure. When you use the internet or other digital devices, you might share information about yourself, like your name, address, age, or even your interests and hobbies. This information is valuable and can be used by other people or companies to try and sell you things, or even to harm you.

To protect your privacy, there are some simple things you can do:

* Keep your passwords private. Don't share your passwords with anyone, and don't use the same password for all your accounts.

* Think before you share. Before you post anything online, ask yourself if it's something you'd be comfortable with anyone seeing.

* Use privacy settings. Most social media and other digital services have privacy settings that you can use to control who can see your information.

* Be careful with personal information. Don't give out your personal information, like your name, address, or phone number, to people or companies you don't trust.

Remember, it's important to protect your privacy in the digital world to stay safe and secure online.
